```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### For Machine A

```python
A = pd.read_csv('task1-1A.csv')
A['支付时间'] = pd.to_datetime(A['支付时间'], errors='coerce', format='%Y/%m/%d %H:%M')
A = A.dropna(subset=['支付时间'])
A.sample(10)
```

```python
A_4 = A[(A.支付时间.dt.year == 2017) & (A.支付时间.dt.month ==  4)]
A_4
```

```python
#交易额
amount = A_4.实际金额.sum()
```

```python
A_4.订单号.value_counts()
```

```python
#订单数量（no duplicate order）
order_num = A_4.shape[0]
```

```python
summary_A = pd.DataFrame({
    'Total Transaction Amount for A (April 2017)': [amount],
    'Total Number of Orders for A (April 2017)': [order_num]
})

summary_A
```
