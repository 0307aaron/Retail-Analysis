```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('merged_data')
data['支付时间'] = pd.to_datetime(data['支付时间'], errors='coerce', format='%Y/%m/%d %H:%M')
```

```python
data['Year'] = data['支付时间'].dt.year.astype('Int64')
data['Month'] = data['支付时间'].dt.month.astype('Int64')
data['Day'] = data['支付时间'].dt.day.astype('Int64')
```

```python
data.sample(10)
```

```python
grouped_data = data.groupby(['Year','Month', '设备ID']).agg(
    total_amount=('实际金额', 'sum'),
    total_orders=('订单号', 'count'),
    unique_days=('支付时间', lambda x: x.dt.date.nunique())
).reset_index()
```

```python
grouped_data.head(5)
```

```python
grouped_data['average_transaction_amount_per_order'] = grouped_data['total_amount'] / grouped_data['total_orders']
grouped_data['average_daily_order_count'] = grouped_data['total_orders'] / grouped_data['unique_days']
```
